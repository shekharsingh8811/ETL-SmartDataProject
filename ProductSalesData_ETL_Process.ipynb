# Creating Star Schema

```python
import pymysql

db = pymysql.connect(host="localhost",    # your host, usually localhost
                     user="root",         # your username
                     password="Mysql@987") # name of the data base  
```

```python
cursor=db.cursor() 
```

```python
cursor.execute("CREATE DATABASE ProductSales")
db.commit()
```

```python
cursor.execute("USE ProductSales")
db.commit()
```

```python
cursor.execute("""CREATE TABLE Employee
(Employee_ID INT NOT NULL PRIMARY KEY,
Employee_Name VARCHAR(255) NULL,
Employee_Middle VARCHAR(255) NULL,
Employee_LastName VARCHAR(255) NULL)""")

cursor.execute("""CREATE TABLE Customer
(Customer_ID INT NOT NULL PRIMARY KEY,
Customer_Name VARCHAR(255) NULL,
Customer_Middle VARCHAR(255) NULL,
Customer_LastName VARCHAR(255) NULL)""")

cursor.execute("""CREATE TABLE Product
(Product_ID INT NOT NULL PRIMARY KEY,
Product_Name VARCHAR(255) NULL,
Product_Price FLOAT NULL)""")

cursor.execute("""CREATE TABLE Sales
(Sales_ID INT NOT NULL,
Sales_Price FLOAT NULL,
Employee_ID INT NULL,
Customer_ID INT NULL,
Product_ID INT  NULL,
Quantity INT NULL,
FOREIGN KEY(Employee_ID) REFERENCES Employee(Employee_ID),
FOREIGN KEY(Customer_ID) REFERENCES Customer(Customer_ID),
FOREIGN KEY(Product_ID) REFERENCES Product(Product_ID))""")

db.commit()
```
